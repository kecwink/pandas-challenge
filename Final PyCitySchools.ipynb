```python
import pandas as pd
```

```python
#import files from resources
schools_data_path = 'schools_complete.csv'
student_data_path = 'students_complete.csv'
```

```python
#read and display the data 
schools_df = pd.read_csv('schools_complete.csv')
students_df = pd.read_csv('students_complete.csv')
#students_df.head()

#merge the info into one table
school_data_complete = pd.merge(schools_df, students_df, how="left", on=["school_name", "school_name"])
school_data_complete
```

```python
#calculate the number of school analyzed
school_count= school_data_complete['School ID'].value_counts()
number_of_schools = len(school_count)
print(number_of_schools)
```

```python
#calculate the total number of students
student_count = school_data_complete['Student ID'].value_counts()
number_of_students = len(student_count)
print(number_of_students)
```

```python
#calculate the total budget 
school_budgets =school_data_complete['budget'].unique()
#print(school_budgets)
total_budget =school_budgets.sum()
print(total_budget)
```

```python
#calculate the average math score
avg_math_score =round(school_data_complete['math_score'].mean())

print(avg_math_score)
```

```python
#calculate the average reading score
avg_reading_score =round(school_data_complete['reading_score'].mean())

print(avg_reading_score)
```

```python
#calculate overall passing score
passing_rate = ((avg_math_score + avg_reading_score)/2)
print(passing_rate)
```

```python
#Calculate the percentage of students with a passing math score (70 or greater)
passing_math_students = school_data_complete.loc[school_data_complete['math_score'] >= 70]
#print(passing_math_students)
num_passing_math_students = len(passing_math_students)
#print(num_passing_math_students)
percent_passing_math = round((num_passing_math_students/number_of_students)*100)
print(percent_passing_math)
#reduced_kickstarter = narrowed_kickstarter.loc[narrowed_kickstarter["pledged"]>0]
```

```python
#Calculate the percentage of students with a passing reading score (70 or greater)
passing_reading_students = school_data_complete.loc[school_data_complete['reading_score'] >= 70]
#print(passing_reading_students)
num_passing_reading_students = len(passing_reading_students)
#print(num_passing_reading_students)
percent_passing_reading = round((num_passing_reading_students/number_of_students)*100)
print(percent_passing_reading)
```

```python
#create a dataframe that summarizes district statistics
district_summary = pd.DataFrame ({'Total Number of Schools' :number_of_schools, 'Total Number of Students' :number_of_students,'Total Budget' :total_budget,'Average Math Score' :avg_math_score,'Average Reading Score' :avg_reading_score,'% Passing Math' :percent_passing_math,'% Passing Reading' :percent_passing_reading,'% Overall Passing' :passing_rate},index = [0])
district_summary
```

```python
#clean the table
clean_school_summary = school_summary_df.dropna(how='any')
```

```python
#convert size to numeric types
converted_summary = clean_school_summary.copy()
converted_summary["size"] = converted_summary.loc[:, "size"].astype(float)
```

```python
#convert budget to numeric types
converted_summary = clean_school_summary.copy()
converted_summary["budget"] = converted_summary.loc[:, "budget"].astype(float)
```

```python
#drop irrelevant columns
narrowed_summary = converted_summary[['school_name', 'type', 'size', 'budget', 'grade', 'reading_score', "math_score"]]
narrowed_summary
```

```python
#change the school name to the index
narrowed_summary = narrowed_summary.set_index('school_name')
narrowed_summary
```

```python
# hhs= narrowed_summary.loc['Huang High School']
# hhs
```

```python
narrowed_summary['Per Student Budget'] = ((narrowed_summary['budget']) /(narrowed_summary['size']))  
narrowed_summary
```

```python
# create bins for passing math grades and create a new column for passing math grades 
bins = [0, 69, 101]
names = ['0',"1"]
narrowed_summary["Passing Math"] = pd.cut(narrowed_summary["math_score"], bins, labels=names)
narrowed_summary
```

```python
# create bins for passing reading grades and create a new column for passing reading grades 
bins = [0, 69, 101]
names = ['0',"1"]
narrowed_summary["Passing Reading"] = pd.cut(narrowed_summary["reading_score"], bins, labels=names)
narrowed_summary
```

```python
grouped_by_school = narrowed_summary.groupby("school_name")
students_passing = grouped_by_school.first()
students_passing
```

```python

students_passing['Passing Reading'] = students_passing.loc[:, 'Passing Reading'].astype(float)
students_passing['% Passing Reading'] = ((students_passing['Passing Reading'].sum()) /(students_passing['size'])*100)  
students_passing
```

```python
students_passing['Passing Math'] = students_passing.loc[:, 'Passing Math'].astype(float)

students_passing['% Passing Math'] = ((students_passing['Passing Math'].sum()) /(students_passing['size']))  
students_passing
```

```python
#overall passing rate
students_passing['% Overall Passing'] = ((students_passing['% Passing Math']) +(students_passing['% Passing Reading'])/2)  
students_passing
```

```python
students_passing['Average Math Score'] = (students_passing['math_score'].sum()/ students_passing['size'])
students_passing
```

```python
students_passing['Average Reading Score'] = (students_passing['reading_score'].sum()/ students_passing['size'])
students_passing
```

```python
#keep relevant columns
clean_schools = students_passing[[ 'type', 'size','budget', 'Per Student Budget', '% Passing Reading', '% Passing Math', '% Overall Passing', 'Average Math Score', 'Average Reading Score']]
clean_schools
```
